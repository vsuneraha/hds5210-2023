## #04 -

Which of those State values has the best average performance on the `Star Rating for health team communicated well with them` score?

```python
# Group by State and calculate the average score for 'Star Rating for health team communicated well with them'
state_avg_scores = hhcaps.groupby('State')['Star Rating for health team communicated well with them'].mean()

# Find the State with the best average performance
best_state = state_avg_scores.idxmax()

print("State with the best average performance:", best_state)
```

## #05 -

What was the average score on `Star Rating for how patients rated overall care from agency` for providers listed as having a `Type of Ownership` of `Hospital Based Program`

**NOTE: Enter your answer with two decimal precision: 0.00**

```python
# Filter the DataFrame for providers with Type of Ownership 'Hospital Based Program'
hospital_based_program_df = hhcaps[hhcaps['Type of Ownership'] == 'Hospital Based Program']

# Calculate the average score for 'Star Rating for how patients rated overall care from agency'
average_score = hospital_based_program_df['Star Rating for how patients rated overall care from agency'].mean()

# Print the result with two decimal precision
print(f"Average score on Star Rating for overall care from agency for Hospital Based Programs: {average_score:.2f}")
```
